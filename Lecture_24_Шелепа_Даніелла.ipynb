```python
from google.colab import auth
import google.auth
from google.cloud import bigquery
from oauth2client.client import GoogleCredentials

auth.authenticate_user()
project_id = 'firm-reef-477813-j3'
bigquery_client = bigquery.Client(project = project_id)


dataset_id = 'junior_da_py_personal'
table_id = 'shopping_data'

query = (
    'SELECT * '
    'FROM `firm-reef-477813-j3.junior_da_py_personal.d_shopping_data` '
)

query_job = bigquery_client.query(query)

shopping_data = query_job.result().to_dataframe()
shopping_data.head()
```

```python
customers_data = customers_data.drop_duplicates()
```

```python
shopping_data = shopping_data.drop_duplicates()
```

```python
import pandas as pd
unioned_stat = pd.merge(shopping_data, customers_data, how='left', on="customer_id")
# unioned_stat['price'] = unioned_stat['price'].astype(float)
unioned_stat['price'] = unioned_stat['price'].replace(',', '.', regex=True).astype(float)
unioned_stat['category'].fillna("UNKNOWN", inplace=True)
filtered_stat = unioned_stat.dropna(subset=["shopping_mall"])
unioned_stat['invoice_date'] = pd.to_datetime(unioned_stat['invoice_date'])

unioned_stat
```
